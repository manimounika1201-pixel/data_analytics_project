```python
import pandas as pd
df=pd.read_csv("Salary.csv")
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
sns.countplot(x='Gender',data=df,color="red")
plt.title("Gender")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x='Gender',y="Salary",data=df,color="blue")
plt.title("salary details")
plt.xlabel("Gender")
plt.ylabel("Salary")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.histplot(df["Education Level"],bins=20,kde=True,color="red")
plt.title("salary details")
plt.xlabel("Education Level")
plt.ylabel("Country")
plt.show()
```

```python
Education_Level_counts=df["Education Level"].value_counts()
plt.figure(figsize=(8,4))
plt.pie(Education_Level_counts,labels=["class1","class2","class3","class0"],colors=["red","yellow","green","blue"],autopct="%1.1f%%")
plt.title("Education Level")
plt.show()
```
